```python
import pandas as pd
import numpy as np

df = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
df.head()
```

```python
print ('Number of Rows :', df.shape[0])
print ('Number of Columns :', df.shape[1])
```

```python
df.describe()
```

### 1.2 Splitting data in train and test
Here we are using sklearn for this 

```python
from sklearn.model_selection import train_test_split
X = df.to_numpy()[:,0:8] 
Y = df.to_numpy()[:,8]
seed = 42
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = seed)
print (f'Shape of Train Data : {X_train.shape}')
print (f'Shape of Test Data : {X_test.shape}')
```

## Step 2: Define the Model
* Models in Keras are defined as a sequence of layers. We create a `Sequential` model and add layers one at a time. 
*  Fully connected layers are defined using the `Dense` class. We can specify the number of neurons in the layer as the first argument, specify the activation function using the `activation` argument

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential([
    Dense(24, input_dim = (8), activation = 'relu'),
    Dense(12, activation = 'relu'),
    Dense(1, activation = 'sigmoid'),
])
```

## Step 3: Compile the model
We specify the cost/loss function to the datset. We will use `binary_crossentropy` cost function that is basically used for sigmoid type of functions which is defined in Keras for binary classification problem. We also specify the optimizer used to search through different weights for the network. Here we will use `adam` optimizer

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

### Step 4: Fit Model
We can fit the model using the `fit()` method. 
batch_size for setting the number of instances that are evaluated before a weight update in the network

```python
history = model.fit(X_train, y_train, epochs=150, batch_size=32, verbose = 1)
```

### Step 5: Evaluate model
We can evaluate our model on validation dataset using the `evaluation()` function. 

```python
scores = model.evaluate(X_test, y_test)
print (f'{model.metrics_names[1]} : {round(scores[1]*100, 2)} %')
```
